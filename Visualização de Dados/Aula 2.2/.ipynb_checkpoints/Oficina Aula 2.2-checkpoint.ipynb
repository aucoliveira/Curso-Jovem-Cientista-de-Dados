1. Importe as bibliotecas pandas e matplotlib, usando a palavra mágica inline para exibir as visualizações no notebook; e carregue os dados, que estão disponibilizados neste link:dados, em um DataFrame. Assim, você pode carregar os dados pela URI ou salvar em sua máquina e carregá-los localmente;

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
games = pd.read_csv('https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro04/oficina/arquivos/dadosoficina04.txt')
games.head()
```

2. Crie uma visualização que mostre a proporção de vendas global entre gêneros de jogos da empresa Nintendo;

```python
nintendo= games[games['Publisher'] == 'Nintendo']
nintendo_generos = nintendo.groupby(by='Genre').sum()
nintendo_generos
plt.title('Proporção de vendas de jogos da Nintendo  por gênero')
plt.pie(x=nintendo_generos['Global_Sales'], labels = nintendo_generos.index, autopct='%1.1f%%',startangle=90, pctdistance=0.85)
plt.show()
```

```python
porcentagens = nintendo_generos['Global_Sales']*100.0/nintendo_generos['Global_Sales'].sum()
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(15,5))
plt.bar(nintendo_generos.index,porcentagens)
plt.title('Proporção de vendas de jogos da Nintendo  por gênero')
plt.show()
```

3. Crie uma visualização que mostre a proporção de quantidade de jogos por plataforma a partir do ano de 2015.

```python
ano= games[games['Year'] >= 2015.0]
ano_plataforma = ano.groupby(by='Platform').count()
porcentagens = ano_plataforma['Year']*100.0/ano_plataforma['Year'].count()
plt.figure(figsize=(15,5))
plt.bar(ano_plataforma.index, porcentagens)
plt.title('Proporção de jogos por plataforma de 2015 até o momento')
plt.show()
```
