1. Importe as bibliotecas pandas, matplotlib, seaborn e use o comando mágico %matplotlib inline para exibir as visualizações no notebook. Carregue os dados em um DataFrame. Os dados estão disponibilizados no link: dados. Você pode carregar os dados pela URI ou salvar em sua máquina e carregá-los localmente.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

filmes_uri = 'https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro03/oficina/arquivos/dadosoficina03.txt'
filmes = pd.read_json(filmes_uri)
filmes.head()
```

2.Crie uma visualização que mostre o orçamento de todos os filmes do conjunto de dados. Qual o filme com maior orçamento e com menor orçamento?

```python
plt.figure(figsize=(20,15))
plt.title('Visualização do orçamento de todos os filmes')
sns.barplot(x="Worldwide_Gross_M",y='Film', color ='r', data=filmes)
plt.show()


maior_orcamento = filmes.groupby(by='Worldwide_Gross_M').max()
maior_orcamento
menor_orcamento = filmes.groupby(by='Worldwide_Gross_M').min()
menor_orcamento
```

3.Crie uma visualização que mostre a bilheteria de todos os filmes do conjunto de dados. Qual o filme com maior bilheteria e com menor bilheteria?

```python
plt.figure(figsize=(20,15))
plt.title('Visualização da bilheteria de todos os filmes')
sns.barplot(x="Budget_M",y='Film', color ='r', data=filmes)
plt.show()
```

```python
menor_bilheteria = filmes.groupby(by='Budget_M').min()
menor_bilheteria
```

4. Crie uma visualização que mostre a quantidade de filmes por gênero em cada ano. Em que anos foram feitos menos filmes de ação? E em que ano foram feitos menos filmes de animação?

```python
plt.figure(figsize=(15,8))
plt.title('Filmes por gênero em cada ano')
sns.countplot(x="Year", hue= 'Genre', data=filmes)
plt.show()
```

5. Crie uma visualização que mostre a média de orçamento de filmes por gênero em cada ano. Em que ano os filmes de ação tiveram maior média de orçamento? E em que ano os filmes de animação tiveram a maior média de orçamento?

```python
plt.figure(figsize=(15,8))
plt.title('Média do orçamento de filmes por gênero em cada ano')
sns.barplot(x="Year", y = 'Worldwide_Gross_M' , ci = None, hue= 'Genre', data=filmes)
plt.show()
```
